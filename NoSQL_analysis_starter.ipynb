```python
# Import dependencies
from pymongo import MongoClient
from pprint import pprint
import pandas as pd
```

```python
# Create an instance of MongoClient
mongo = MongoClient(port=27017)
```

```python
# assign the collection to a variable
establishments = db['establishments']
```

### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
hygiene_20_query = {"scores.Hygiene": 20}


# Use count_documents to display the number of documents in the result
number_of_hygiene_20 = establishments.count_documents(hygiene_20_query)
print(f"The number of documents with a hygiene score equal to 20 is {number_of_hygiene_20}.")

# Display the first document in the results using pprint
hygiene_20_results = list(establishments.find(hygiene_20_query))
pprint(hygiene_20_results[0])
```

```python
# Convert the result to a Pandas DataFrame
hygiene_20_df = pd.json_normalize(hygiene_20_results)
# Display the number of rows in the DataFrame
print(f"The number of rows in the DataFrame is {len(hygiene_20_df)}")
# Display the first 10 rows of the DataFrame
hygiene_20_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
London_rating_gte_4_query = {'LocalAuthorityName': {"$regex": "London"}, "RatingValue": {"$gte" : 4}}

# Use count_documents to display the number of documents in the result
number_of_London_rating_gte_4 = establishments.count_documents(London_rating_gte_4_query)
print(f"The number of documents with a hygiene score equal to 20 is {number_of_London_rating_gte_4}.")

# Display the first document in the results using pprint
London_rating_gte_4_results = list(establishments.find(London_rating_gte_4_query))
pprint(London_rating_gte_4_results[0])
```

```python
# Convert the result to a Pandas DataFrame
London_rating_gte_4_df = pd.json_normalize(London_rating_gte_4_results)

# Display the number of rows in the DataFrame
print(f"The number of rows in the DataFrame is {len(London_rating_gte_4_df)}")

# Display the first 10 rows of the DataFrame
London_rating_gte_4_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score
new_restaurant_result= establishments.find_one({'BusinessName': "Penang Flavours"})

degree_search = 0.01
latitude = new_restaurant_result['geocode']['latitude']
longitude = new_restaurant_result['geocode']['longitude']

query = {"RatingValue": 5, "$or": [{"geocode.latitude": {"$gte": (latitude - degree_search), "$lte": (latitude + degree_search)}}, {"geocode.longitude": {"$gte": (longitude - degree_search), "$lte": (longitude + degree_search)}}]}
sort =  [("scores.Hygiene", 1)]

# Print the results
top_5_rating_5_close_to_Penang_Flavours_results = list(establishments.find(query).sort(sort).limit(5))
pprint(top_5_rating_5_close_to_Penang_Flavours_results)
```

```python
# Convert result to Pandas DataFrame
top_5_rating_5_close_to_Penang_Flavours_df = pd.json_normalize(top_5_rating_5_close_to_Penang_Flavours_results)
top_5_rating_5_close_to_Penang_Flavours_df
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest
match_query = {"$match": {'scores.Hygiene': 0}}

group_query = {"$group": {"_id": "$LocalAuthorityName",
                          "count": {"$sum": 1}}}

sort_query = {"$sort": {"count": -1}}

pipeline = [match_query, group_query, sort_query]
# Print the number of documents in the result
results = list(establishments.aggregate(pipeline))
print(f"The number of documents in the result is {len(results)}")
# Print the first 10 results
pprint(results[0:10])
```

```python
# Convert the result to a Pandas DataFrame
group_by_Authority_has_0_hygiene_score_df = pd.DataFrame(results)
# Display the number of rows in the DataFrame
print(f"The number of rows in the DataFrame is {len(group_by_Authority_has_0_hygiene_score_df)}")
# Display the first 10 rows of the DataFrame
group_by_Authority_has_0_hygiene_score_df.head(10)
```
